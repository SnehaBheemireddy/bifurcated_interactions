```python
import urllib.parse
import urllib.request
import requests

#columns = ['DR','PDB','ID','Method','Resolution','A','Chain','length']
search_value = 'PDB;'
def get_pdb_data(url, search_value):
    data_lst = []
    response = requests.get(url).content.decode('utf-8')

    for ln in response.split('\n'):
        if search_value in ln:
            print(ln)
            ln_data = [i.split('=') for i in ln.split() if '=' in i]
            data_lst.append(ln.split()[0:-1]+ln_data[0])
    df = pd.DataFrame(data_lst)
    return df
for i in list_uniprot:
    url = 'https://www.uniprot.org/uniprot/'+i+'.txt'
    ab = get_pdb_data(url, search_value)
    ab.to_csv(i+'_uni.csv')
```
